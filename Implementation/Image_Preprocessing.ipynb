```python
#Importing necessary libraries
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
#reading the original image using OpenCV
image = cv2.imread('53.jpg')
cv2.imshow('Original',image)
cv2.waitKey(0)
RGB_im = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)  #opencv reads image in BGR format to display in matplot lib we convert it to RGB
plt.imshow(RGB_im)
```

```python
#Converting the colored image to greyscale
gray = cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
cv2.imshow('RBG2GRAY',gray)
cv2.imwrite('RBG2GRAY.jpg', gray)
cv2.waitKey(0)
RGB_im = cv2.cvtColor(gray, cv2.COLOR_BGR2RGB) #opencv reads image in BGR format to display in matplot lib we convert it to RGB
plt.imshow(RGB_im)
```

```python
#applying median filter for Salt and pepper/impulse noise
filter1 = cv2.medianBlur(gray,5)
cv2.imshow('Median FIlter',filter1)
cv2.imwrite('Median_Filter.jpg', filter1)
cv2.waitKey(0)
RGB_im = cv2.cvtColor(filter1, cv2.COLOR_BGR2RGB) #opencv reads image in BGR format to display in matplot lib we convert it to RGB
plt.imshow(RGB_im)
```

```python
#applying gaussian blur to smoothen out the image edges
filter2 = cv2.GaussianBlur(filter1,(5,5),0)
cv2.imshow('Gaussuan Filter',filter2)
cv2.imwrite('Gaussian_Filter.jpg', filter2)
cv2.waitKey(0)
RGB_im = cv2.cvtColor(filter2, cv2.COLOR_BGR2RGB) #opencv reads image in BGR format to display in matplot lib we convert it to RGB
plt.imshow(RGB_im)
```

```python
#applying non-localized means for final Denoising of the image
dst = cv2.fastNlMeansDenoising(filter2,None,17,9,17)
cv2.imshow('Denoised',dst)
cv2.imwrite('Denoised.jpg', dst)
cv2.waitKey(0)
RGB_im = cv2.cvtColor(dst, cv2.COLOR_BGR2RGB) #opencv reads image in BGR format to display in matplot lib we convert it to RGB
plt.imshow(RGB_im)
```

```python
#converting the image to binarized form using adaptive thresholding
th1 = cv2.adaptiveThreshold(dst,255,cv2.ADAPTIVE_THRESH_GAUSSIAN_C,cv2.THRESH_BINARY,11,2)
cv2.imshow('Final',th1)
cv2.imwrite('Final.jpg', th1)
cv2.waitKey(0)
RGB_im = cv2.cvtColor(th1, cv2.COLOR_BGR2RGB) #opencv reads image in BGR format to display in matplot lib we convert it to RGB
plt.imshow(RGB_im)
```
